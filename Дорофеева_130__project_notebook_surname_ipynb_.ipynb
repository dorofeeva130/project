```python
def user_score_group(user_score):

    if user_score <= 3:
        return '0.0-3.0'
    if user_score <= 5:
        return '3.1-5.0'
    if user_score <= 7:
        return '5.1-7.0'
    return '7.1-10.0'
```

Протестируем работу функции для каждого правила. Проверим, в какую категорию попадёт та или иная оценка от пользователей. Чаще всего перед покупкой игры люди больше замечают их отзывы, чем оценки критиков.

```python
print(user_score_group(2.5))
```

```python
print(user_score_group(5.2))
```

```python
print(user_score_group(9.9))
```

Написанная функция работает корректно. Далее создаем сам столбец.

```python
games['user_score_group'] = games['user_score'].apply(user_score_group)

games.sample(5)
```

**2.8.3.** Добавьте столбец, используя функцию для одной строки.

```python
def popular_game(row):

    user_score_group = row['user_score_group']

    if user_score_group == '7.1-10.0':
        return 'popular'
    return '-'
```

#### **5.1** Отбор и выделение признаков

```python
# отберем признаки с наиболее высокой корреляцией с целевой переменной
# и поместим их в переменную feature

features = games[['year_of_release', 'na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'critic_score']]
```

#### **5.2** Platform, Genre, Rating

```python
# целевую переменную поместим в переменную target

target = games['user_score']
```

```python
print(type(features), type(target))
```

```python
# разобьем данные на обучающую и тестовую выборку
# размер тестовой выборки составит 30%
# также зададим точку отсчета для воспроизводимости

features_train, features_test, target_train, target_test = train_test_split(
    features, target,
    test_size=0.3,
    random_state=RANDOM_STATE
)
```

```python
# посмотрим на новую размерность обучающей выборки

print(f'Признаки обучающей выборки: {features_train.shape}')
print(f'Целевой признак обучающей выборки: {target_train.shape}')
```

```python
# и тестовой выборки

print(f'Признаки тестовой выборки: {features_test.shape}')
print(f'Целевой признак тестовой выборки: {target_test.shape}')
```

#### **5.3** Обучение модели

```python
# создадим объект этого класса и запишем в переменную model

%%time

model = LinearRegression()
```

```python
# обучим модель линейной регрессии

%%time

model.fit(features_train, target_train)
```

```python
# на основе нескольких независимых переменных (feature) предскажем пользовательскую оценку (target)

target_pred = model.predict(features_test)
```

```python
# выведем первые пять значений с помощью диапазона индексов

print(target_pred[:5])
```

```python
target_test
```

```python
coefficients = pd.DataFrame({'features': features.columns, 'coefficients': model.coef_})
print(coefficients)
```

#### **5.4** Оценка качества модели

```python
# выведем корень среднеквадратической ошибки
# сравним тестовые и прогнозные значения оценок пользователей

print('Root Mean Squared Error {RMSE:}', np.sqrt(metrics.mean_squared_error(target_test, target_pred)))
```

```python
# рассчитаем коэффициент детерминации **R2**

print('R2:', np.round(metrics.r2_score(target_test, target_pred), 2))
```

**Вывод** <br>
* R2 = 0.7. Следовательно, **70% результатов** от модели **совпадут с ожидаемым результатом.**
